```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Read the CSV file into a pandas DataFrame
data = pd.read_csv('/content/sparktime.csv')

# Convert the 'time' column to datetime format
data['Time'] = pd.to_datetime(data['Time'])

# Sort the DataFrame by the 'time' column
data.sort_values(by='Time', inplace=True)

# Create a figure and axis
fig, ax = plt.subplots()

# Plot a vertical line for each person's time
for idx, row in data.iterrows():
    ax.axvline(x=row['Time'], label=row['name'])

# Format the x-axis as dates
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M:%S'))
plt.xticks(rotation=45)

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Persons')
plt.title('Timeline of Persons')

# Add legend
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data_file = '/content/fspark.csv'
df = pd.read_csv(data_file)

# Load the CSV file containing roll numbers, departments, and student names
info_file = '/content/basic details.csv'
info_df = pd.read_csv(info_file)

# Get the student's name from the user
student_name = input("Enter the name of the student: ")

# Filter data for the specified student's name
student_data = df[df['Name'] == student_name]

if student_data.empty:
    print(f"No data found for student: {student_name}")
else:
    # Remove duplicates based on the same timestamp
    student_data = student_data.drop_duplicates(subset=['Timestamp'])
    sorted_data = student_data.sort_values('Timestamp')

    # Merge student data with info data based on student's name
    merged_data = sorted_data.merge(info_df, on='Name', how='left')

    print("Student Information for the Full Day:")
    for _, row in merged_data.iterrows():
        print(f"Time: {row['Timestamp']}")
        print(f"Name: {row['Name']}")
        print(f"Location: {row['Camera']}")
        print(f"Roll No: {row['rollno']}")
        print(f"Department: {row['dept']}")
        print()

    plt.figure(figsize=(10, 6))
    plt.plot(merged_data['Timestamp'], merged_data['Camera'], marker='o')
    plt.xlabel('Time')
    plt.ylabel('Location')
    plt.title('Timeline Chart')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data_file = '/content/fspark.csv'
df = pd.read_csv(data_file)

# Load the CSV file containing roll numbers, departments, and student names
info_file = '/content/basic details.csv'
info_df = pd.read_csv(info_file)

# Get the student's roll number from the user
student_roll = input("Enter the roll number of the student: ")

# Get the date in the format "dd-mm-yyyy" from the user
input_date = input("Enter the date (dd-mm-yyyy): ")

# Filter info data for the specified roll number
student_info = info_df[info_df['rollno'] == student_roll]

if student_info.empty:
    print(f"No information found for student with Roll No: {student_roll}")
else:
    student_name = student_info.iloc[0]['Name']

    # Filter data for the specified student's name
    student_data = df[df['Name'] == student_name]

    if student_data.empty:
        print(f"No data found for student: {student_name}")
    else:
        # Replace values in 'Camera' column
        student_data['Camera'] = student_data['Camera'].replace({'camera1': 'AI DEPT', 'camera2': 'ADMIN BLOCK'})

        # Extract the date and time from the Timestamp column
        student_data['Timestamp'] = pd.to_datetime(student_data['Timestamp'])
        student_data['Date'] = student_data['Timestamp'].dt.strftime('%d-%m-%Y')
        student_data['Time'] = student_data['Timestamp'].dt.strftime('%H:%M:%S')

        # Filter data based on the input date
        filtered_data = student_data[student_data['Date'] == input_date]

        if filtered_data.empty:
            print(f"No data found for student: {student_name} on {input_date}")
        else:
            sorted_data = filtered_data.sort_values('Timestamp')

            print(f"Student Information for {input_date}:")
            for _, row in sorted_data.iterrows():
                print(f"Time: {row['Time']}")
                print(f"Name: {row['Name']}")
                print(f"Location: {row['Camera']}")
                print()

            # Optionally, you can also display the roll number and department here
            print(f"Roll No: {student_roll}")
            print(f"Department: {student_info.iloc[0]['dept']}")

            plt.figure(figsize=(10, 6))
            plt.plot(sorted_data['Time'], sorted_data['Camera'], marker='o')
            plt.xlabel('Time')
            plt.ylabel('Location')
            plt.title(f'Timeline Chart for Student {student_name} on {input_date}')
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()
```
